```python
def arithmetic_sum(a, d, n):
    return n / 2 * (2 * a + (n - 1) * d)

a = 1
d = 2
n = 10

sum_progression = arithmetic_sum(a, d, n)
print(sum_progression)
```

```python
def extract_digits(number):
    if 10 <= number <= 99:
        left_digit = number // 10
        right_digit = number % 10
        return left_digit, right_digit
    else:
        return "Error: Not a two-digit number"

number = 42
digits = extract_digits(number)
print(digits)
```

```python
def king_can_move(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    return dx <= 1 and dy <= 1

x1, y1 = 4, 4
x2, y2 = 5, 5

if king_can_move(x1, y1, x2, y2):
    print("Король может перейти за один ход.")
else:
    print("Король не может перейти за один ход.")
```

```python
#3)
def is_parallelogram(x1, y1, x2, y2, x3, y3, x4, y4):
    #Достаточно проверить, что диагонали делят друг друга пополам.
    mid_AC = ((x1 + x3) / 2, (y1 + y3) / 2)
    mid_BD = ((x2 + x4) / 2, (y2 + y4) / 2)

    return mid_AC == mid_BD
x1, y1, x2, y2, x3, y3, x4, y4 = float(input()), float(input()), float(input()), float(input()), float(input()), float(input()), float(input()), float(input())


if is_parallelogram(x1, y1, x2, y2, x3, y3, x4, y4):
    print("Points form a parallelogram.")
else:
    print("Points don't form a parallelogram.")
```

```python
#4)
def compare_distances(km, feet):
    km_to_m = km * 1000
    feet_to_m = feet * 0.305

    if km_to_m < feet_to_m:
        return "Distance in km is shorter."
    elif km_to_m > feet_to_m:
        return "Distance in feet is shorter."
    else:
        return "Distances are equal."

km_distance = float(input())
feet_distance = float(input())
result = compare_distances(km_distance, feet_distance)
print(result)
```

```python
#1
def reverse_num(n):
  new_n = 0
  while n != 0:
    new_n += n%10
    new_n *= 10
    n//=10
  new_n //=10
  return new_n
x = int(input())
print(reverse_num(x))
```

```python
#2
def gcd(a, b):
  if (b == 0):
    return a
  return gcd(b, a%b)
def gcd_of_four(a, b, c, d):
  return gcd(gcd(a, b), gcd(c, d))
x, y, z, w = int(input()), int(input()), int(input()), int(input())
print(gcd_of_four(x, y, z, w))
```

```python
#3
def C(m, n):
  if m > n:
    return 0
  if m == 0 or m == n:
    return 1
  return C(m, n-1) + C(m-1, n-1)
m, n = int(input()), int(input())

print(C(m, n))
```
