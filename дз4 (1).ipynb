```python
#
#1
from random import random
f = open("newfile.txt", "w")
for i in range(10):
  f.write(str(int(random() * 100) + 1))
  f.write("\n")
f.close()
g = open("g.txt", "w")
f = open("newfile.txt", "r")
for i in f:
  k = int(i[:len(i) - 1])
  if k % 2 == 0:
    g.write(i)
f.close()
g.close()
with open("g.txt") as g:
  for i in g:
    print(i,end= '')
```

```python
#2
# int, float
import pickle
from random import randint, random

l = [[randint(1, 10), random() * 25 + 1] for i in range(10)]

with open("bagazh.bin", "wb") as f:
    pickle.dump(l, f)

sm = 0
weight = 0

with open("bagazh.bin", "rb") as f:
    k = pickle.load(f)
    for i in k:
        sm += i[0]
        weight += i[1]

avg = weight / sm
print(avg)
for i in k:
    local_avg = i[1]/i[0]
    if abs(local_avg - avg) <= 0.3:
        print(i[0], i[1])
```

```python
#lesson 4
from random import shuffle
from datetime import datetime

class Tester:
  def __init__(self, file_name):
    self.test = []
    self.q = 0

    f = open(file_name, 'r')
    for i in f:
      self.test.append(i.split('|'))
      self.test[-1][-1] = self.test[-1][int(self.test[-1][-1])]
      #print(self.test[-1])
    f.close()
  def shuffle_test(self):
    shuffle(self.test)
    for i in range(len(self.test)):
      sub_arr = self.test[i][1:5]
      shuffle(sub_arr)
      self.test[i][1:5] = sub_arr
  def ask_question(self):
    if (self.q == len(self.test)):
      return -1
    print(f"  {self.q+1}")
    print(self.test[self.q][0])
    print(" : ")
    for i in range(1, 5):
      print(f"{i}.{self.test[self.q][i]}")
    ans = self.test[self.q][int(input(" : "))]
    if ans == self.test[self.q][-1]:
      print("!")
      self.q += 1
      return 1
    else:
      print(" :(")
      self.q += 1
      return 0

h = Tester("test.txt")
f = open("results.txt", "w")
h.shuffle_test()
cnt, all = 0, 0
fl = 1
start_time = datetime.now()
while (fl != -1):
  fl = h.ask_question()
  cnt += fl
  all += 1
end_time = datetime.now()
f.write(f" : {start_time}")
f.write(f" : {end_time}")
f.write(f"  : {all}")
f.write(f"  : {cnt}")
f.write(f"  : {cnt/10 * 100}%")
f.close()
```
