```python
#1
def word_frequency(text):
    words = text.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    return frequency
print(word_frequency("HELLO HELLO < < < MY NAME IS PAUL"))
```

```python
#2
def invert_dictionary(d):
    inverted = {}
    for key, value in d.items():
        if value in inverted:
            if isinstance(inverted[value], list):
                inverted[value].append(key)
            else:
                inverted[value] = [inverted[value], key]
        else:
            inverted[value] = key
    return inverted
print(invert_dictionary({"h" : 2, "k" : 1, "jj" : 1}))
```

```python
#3
def merge_dictionaries(d1, d2):
    merged = d1.copy()
    for key, value in d2.items():
        merged[key] = merged.get(key, 0) + value
    return merged
print(merge_dictionaries({"a": 1, "b": 2}, {"b": 3, "c": 4}))
print(merge_dictionaries({"x": 10}, {"y": 20}))
```

```python
#4
def filter_by_value(d, threshold):
    return {k: v for k, v in d.items() if v > threshold}
print(filter_by_value({"a": 5, "b": 3, "c": 8}, 4))
print(filter_by_value({"a": 1, "b": 2}, 5))
```

```python
#5
def key_with_max_value(d):
    return max(d, key=d.get)
print(key_with_max_value({"a": 1, "b": 5, "c": 3}))
print(key_with_max_value({"x": 10, "y": 10}))
```

```python
#6
def key_with_min_value(d):
    return min(d, key=d.get)
print(key_with_min_value({"a": 1, "b": 5, "c": 3}))
print(key_with_min_value({"x": 5, "y": 5}))
```

```python
#7
def remove_keys(d, keys_to_remove):
    return {k: v for k, v in d.items() if k not in keys_to_remove}
print(remove_keys({"a": 1, "b": 2, "c": 3}, ["b"]))
print(remove_keys({"name": "John", "age": 30}, ["age"]))
```

```python
#8
def group_keys_by_length(d):
    grouped = {}
    for key in d.keys():
        length = len(key)
        if length not in grouped:
            grouped[length] = []
        grouped[length].append(key)
    return grouped
print(group_keys_by_length({"a": 1, "bb": 2, "ccc": 3, "kkk":1}))
print(group_keys_by_length({"hi": 1, "hello": 2}))
```

```python
#9
def find_keys_by_value(d, value):
    return [k for k, v in d.items() if v == value]
print(find_keys_by_value({"a": 1, "b": 2, "c": 1}, 1))
print(find_keys_by_value({"x": 10, "y": 20}, 15))
```

```python
#10
def combine_dictionaries(d1, d2):
    combined = {}
    all_keys = set(d1.keys()) | set(d2.keys())
    for key in all_keys:
        combined[key] = [d1.get(key), d2.get(key)]
    return combined
print(combine_dictionaries({"a": 1, "b": 2}, {"b": 3, "c": 4}))
# {'a': [1, None], 'b': [2, 3], 'c': [None, 4]}

print(combine_dictionaries({"x": 10}, {"y": 20}))
# {'x': [10, None], 'y': [None, 20]}
```

```python
#1
def list_to_set(lst):
    return set(lst)

print(list_to_set([1, 2, 2, 3, 4, 4, 4]))
print(list_to_set(['a', 'b', 'a', 'c']))
```

```python
def remove_duplicates(lst):
    return list(set(lst))

print(remove_duplicates([1, 2, 2, 3, 3, 3]))
print(remove_duplicates(['x', 'y', 'x', 'z']))
```

```python
#3
def set_intersection(set1, set2):
    return set1 & set2

print(set_intersection({1, 2, 3}, {2, 3, 4}))
print(set_intersection({'a', 'b'}, {'b', 'c'}))
```

```python
#4
def set_union(set1, set2):
    return set1 | set2

print(set_union({1, 2}, {3, 4}))
print(set_union({'apple'}, {'banana'}))
```

```python
#5
def set_difference(set1, set2):
    return set1 - set2

print(set_difference({1, 2, 3}, {2, 4}))
print(set_difference({'a', 'b', 'c'}, {'b'}))
```

```python
#6
def symmetric_difference(set1, set2):
    return set1 ^ set2

print(symmetric_difference({1, 2, 3}, {3, 4, 5}))
print(symmetric_difference({'x', 'y'}, {'y', 'z'}))
```

```python
#7
def is_subset(subset, mainset):
    return subset.issubset(mainset)

print(is_subset({1, 2}, {1, 2, 3}))
print(is_subset({1, 4}, {1, 2, 3}))
```

```python
#8
def is_superset(mainset, subset):
    return mainset.issuperset(subset)

print(is_superset({1, 2, 3}, {1, 2}))
print(is_superset({1, 2}, {1, 2, 3}))
```

```python
#9
def add_elements_to_set(input_set, elements):
    input_set.update(elements)
    return input_set

my_set = {1, 2}
print(add_elements_to_set(my_set, [3, 4]))
print(add_elements_to_set({'a'}, ['b', 'c']))
```

```python
#10
def remove_elements_from_set(input_set, elements):
    input_set.difference_update(elements)
    return input_set

my_set = {1, 2, 3, 4}
print(remove_elements_from_set(my_set, [2, 4]))
print(remove_elements_from_set({'a', 'b', 'c'}, ['b']))
```

```python
#1
def list_to_tuple(lst):
    return tuple(lst)

print(list_to_tuple([1, 2, 3, 4]))
print(list_to_tuple(['a', 'b', 'c']))
```

```python
#2
def combine_tuples(tuple1, tuple2):
    return tuple1 + tuple2

print(combine_tuples((1, 2), (3, 4)))
print(combine_tuples(('x',), ('y', 'z')))
```

```python
#3
def contains_element(tpl, element):
    return element in tpl

print(contains_element((1, 2, 3), 2))
print(contains_element(('a', 'b'), 'c'))
```

```python
#4
def first_index(tpl, element):
    return tpl.index(element)

print(first_index(('a', 'b', 'c'), 'b'))
print(first_index((10, 20, 30, 20), 20))
```

```python
#5
def count_occurrences(tpl, element):
    return tpl.count(element)

print(count_occurrences((1, 2, 2, 3), 2))
print(count_occurrences(('a', 'a', 'b'), 'c'))
```

```python
#6
def max_element(tpl):
    return max(tpl)

print(max_element((10, 5, 25, 15)))
print(max_element(('x', 'y', 'z')))
```

```python
#7
def min_element(tpl):
    return min(tpl)

print(min_element((50, 30, 40, 20)))
print(min_element(('m', 'n', 'o')))
```

```python
#8
def sorted_tuple(tpl):
    return sorted(tpl)

print(sorted_tuple((3, 1, 4, 2)))
print(sorted_tuple(('banana', 'apple')))
```

```python
#9
def split_tuple(tpl):
    mid = len(tpl) // 2
    return tpl[:mid], tpl[mid:]

print(split_tuple((1, 2, 3, 4)))
print(split_tuple(('a', 'b', 'c')))
```

```python
#10
def reverse_tuple(tpl):
    return tpl[::-1]

print(reverse_tuple((1, 2, 3, 4)))
print(reverse_tuple(('x', 'y', 'z')))
```
