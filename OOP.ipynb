```python
class Course:
  def __init__(self, name, program):
    self._name = name
    self._program = program
  def showCourse(self):
    print(self._name, self._program)
class Institute:
  def __init__(self):
    self._courses = []
  def addCourse(self, name, program):
    new_course = Course(name, program)
    self._courses.append(new_course)
  def showCourses(self):
    for i in self._courses:
      i.showCourse()
if __name__ == '__main__':
  vmk = Institute()
  vmk.addCourse('Math', '200h')
  vmk.addCourse('Algebra', '160h')
  vmk.showCourses()
```

```python
from abc import ABC, abstractmethod
class SportEvent(ABC):
  @abstractmethod
  def schedule(self):
    pass
class FootballMatch(SportEvent):
  def __init__(self, date, name):
    self._date = date
    self._name = name
  def schedule(self):
    print("Дата футбольного матча", self._name + ":",  self._date)
class BasketballGame(SportEvent):
  def __init__(self, date, name):
    self._date = date
    self._name = name
  def schedule(self):
    print("Дата баскетбольного матча", self._name + ":", self._date)
if __name__ == '__main__':
  fm = FootballMatch('09.06.2025', 'ЦСКА - СПАРТАК')
  bm = BasketballGame('12.06.2025', 'Golden State vs Lakers')
  fm.schedule()
  bm.schedule()
```

```python
from abc import ABC, abstractmethod
class ITransaction:
  @abstractmethod
  def updateAmount(self, amount):
    pass
  @abstractmethod
  def getTransactionDetails():
    pass
class Transaction(ITransaction):
  def __init__(self, transactionType, amount, date):
    self._transactionType = transactionType
    self._amount = amount
    self._date = date
  def updateAmount(self, amount):
    self._amount = amount
  def getTransactionDetails(self):
    return (self._transactionType, self._amount, self._date)
if __name__ == '__main__':
  transfer = Transaction(1, 121.12, '01.01.25')
  print(transfer.getTransactionDetails())
  transfer.updateAmount(200.1)
  print(transfer.getTransactionDetails())
```
