```python
def arithmetic_sum(a, d, n):
    return n / 2 * (2 * a + (n - 1) * d)

a = 1
d = 2
n = 10

sum_progression = arithmetic_sum(a, d, n)
print(sum_progression)
```

```python
def extract_digits(number):
    if 10 <= number <= 99:
        left_digit = number // 10
        right_digit = number % 10
        return left_digit, right_digit
    else:
        return "Error: Not a two-digit number"

number = 42
digits = extract_digits(number)
print(digits)
```

```python
def bishop_move(x1, y1, x2, y2):
    return abs(x2 - x1) == abs(y2 - y1)

x1, y1 = 1, 1
x2, y2 = 3, 3

valid_move = bishop_move(x1, y1, x2, y2)
print(valid_move)
```
